```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse, urlunparse
from collections import deque
import time

base_url = 'https://www.musashino-u.ac.jp/'
visited_urls = set()
url_title_dict = {}

def scrape_urls_with_queue():
    queue = deque([(base_url, 0)])
    visited_urls.add(base_url)

    while queue:
        url, depth = queue.popleft()

        # 0.5秒待機
        time.sleep(0.5)

        try:
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
            }
            res = requests.get(url, headers=headers, timeout=10)
            res.raise_for_status()
            res.encoding = res.apparent_encoding

            # コンテンツタイプをチェック
            content_type = res.headers.get('Content-Type', '').lower()
            if any(ext in content_type for ext in ['pdf', 'msword', 'excel', 'powerpoint']):
                continue

            if 'text/html' not in content_type:
                continue

            soup = BeautifulSoup(res.text, 'html.parser')
            title = soup.title.string.strip() if soup.title else 'No Title'
            url_title_dict[url] = title
            print(f"[{depth}] {url} -> {title}")

            # 次の階層のリンクを探索
            for link in soup.find_all('a', href=True):
                full_url = urljoin(base_url, link['href'])
                parsed = urlparse(full_url)

                # 同一ドメインのみ許可
                if parsed.netloc == urlparse(base_url).netloc:
                    clean_url = urlunparse(parsed._replace(query='', fragment=''))
                    if clean_url not in visited_urls:
                        visited_urls.add(clean_url)
                        queue.append((clean_url, depth + 1))

        except requests.exceptions.HTTPError as e:
            print(f"HTTPエラー {url}: {e}")
        except requests.exceptions.RequestException as e:
            print(f"リクエストエラー {url}: {e}")
        except Exception as e:
            print(f"エラー {url}: {e}")

print(f"クロール開始: {base_url} (最大深度: {max_depth})")
scrape_urls_with_queue()

print("\n=== URLとタイトルの一覧 ===")
for url, title in sorted(url_title_dict.items()):
    print(f"{url} → {title}")

print(f"\n総ページ数: {len(url_title_dict)}")
```
