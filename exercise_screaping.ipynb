```python
import requests
from bs4 import BeautifulSoup
import time  

url = "https://github.com/trending/python?since=daily"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
repo_items = soup.select("article.Box-row")

for repo in repo_items:
    repo_name = repo.h2.a.text.strip().replace("\n", "").replace(" ", "")
    description_elem = repo.select_one("p")
    description = description_elem.text.strip() if description_elem else ""
    stars_elem = repo.select_one("a[href*='/stargazers']")
    stars = int(stars_elem.text.strip().replace(",", "")) if stars_elem else 0  
    print(f"{repo_name} | {description} | â­ {stars}")


time.sleep(1)  
```
